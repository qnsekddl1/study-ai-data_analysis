### 📝분석 결과 1
📈2023년의 주요 지상파, 공중파 채널에서 방영 되었던 드라마들의 평균 최고 시청률을 분석 해보았다.

1. 대부분의 채널의 평균 최고 시청률이 편차가 상당히 있었다.  
가장 편차가 컸던 KBS2 채널을 본다면 최저 시청률은 '순정복서'의 2.2% 인 반면에 최고 시청률은 '진짜가 나타났다!'의 23.9%  로 표준편차가 7.2 에 달했다.

2. 그에 반면 ENA 채널의 드라마들은 비록 시청률이 낮지만 시청률의 편차가 크지 않은 것으로 나왔다.  
최저 시청률은 '딜리버리 맨!', '보라! 데보라' 모두 1.2%였고 최고 시청률은 '남남'의 5.5%로 표쥰편차가 1.53 밖에 되지않았다.

🤔그렇다면 시청률이 결정 되는 요인에 대해서 분석해 볼 필요가 있어보인다.

#### 시청률에 영향을 미치는 요인에 대해서 첫번째 '유형' 별 차이가 있다는 가설을 세웠다.

```python
kd_df.value_counts('유형')
```

```python
# 토일 방송 드라마들의 목록

kd_df_we = kd_df[kd_df['유형'] == '토일'].reset_index(drop=True)
kd_df_we
```

```python
# 주말 드라마의 평균 최고 시청률

average_rating_we = kd_df_we['최고 시청률'].mean()
print("최고 시청률의 평균:", average_rating_we)
```

```python
# 월화 드라마의 평균 최고 시청률

average_rating_mt = kd_df_mt['최고 시청률'].mean()
print("최고 시청률의 평균:", average_rating_mt)
```

```python
# 수목 방송 드라마들의 목록

kd_df_wt = kd_df[kd_df['유형'] == '수목'].reset_index(drop=True)
kd_df_wt
```

```python
# 수목 드라마의 평균 최고 시청률

average_rating_wt = kd_df_wt['최고 시청률'].mean()
print("최고 시청률의 평균:", average_rating_wt)
```

```python
# 금토 방송 드라마들의 목록

kd_df_fs = kd_df[kd_df['유형'] == '금토'].reset_index(drop=True)
kd_df_fs
```

```python
# 금토 드라마의 평균 최고 시청률

average_rating_fs = kd_df_fs['최고 시청률'].mean()
print("최고 시청률의 평균:", average_rating_fs)
```

```python
# 일일 방송 드라마들의 목록

kd_df_ed = kd_df[kd_df['유형'] == '일일'].reset_index(drop=True)
kd_df_ed
```

```python
import matplotlib.pyplot as plt

# 다섯 개의 데이터프레임에서 '최고 시청률' 열을 가져옵니다.
ratings_we = kd_df_we['최고 시청률']
ratings_mt = kd_df_mt['최고 시청률']
ratings_wt = kd_df_wt['최고 시청률']
ratings_fs = kd_df_fs['최고 시청률']
ratings_ed = kd_df_ed['최고 시청률']

# 그래프 크기 설정
plt.figure(figsize=(10, 6))

# 그래프 그리기
plt.plot(ratings_we.index, ratings_we.values, label='주말드라마')
plt.plot(ratings_mt.index, ratings_mt.values, label='월화드라마')
plt.plot(ratings_wt.index, ratings_wt.values, label='수목드라마')
plt.plot(ratings_fs.index, ratings_fs.values, label='금토드라마')
plt.plot(ratings_ed.index, ratings_ed.values, label='일일드라마')

# 그래프 제목 설정
plt.title('유형별 시청률')

# x축 라벨 설정
plt.xlabel('Index')

# y축 라벨 설정
plt.ylabel('최고 시청률')

# 범례 표시
plt.legend()

# 그래프 보여주기
plt.show()
```

```python
import seaborn as sns

# 상자 그림 그리기
plt.figure(figsize=(10, 6))
sns.boxplot(data=[ratings_we, ratings_mt, ratings_wt, ratings_fs, ratings_ed])
plt.title('연속극 유형별 최고 시청률 분포')
plt.ylabel('최고 시청률')
plt.xticks(ticks=[0, 1, 2, 3, 4], labels=['주말드라마', '월화드라마', '수목드라마', '금토드라마', '일일드라마'])
plt.show()
```
