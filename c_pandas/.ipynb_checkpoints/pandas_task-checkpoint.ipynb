```python
import pandas as pd
import numpy as np

happiness_df = pd.read_csv('https://github.com/qnsekddl1/study-ai/files/14768321/happiness_report_2023.csv')


#1. DataFrmae 확인
display(happiness_df)
```

```python
# DataFrame 중 상위 15개 확인
display(happiness_df.head(15))
```

```python
# Dataframe 중 하위 15개 확인
display(happiness_df.tail(15))
```

```python
# 국가들의 각 척도의 평균, 표준편차, 최소값, 최대값, 사분위를 확인하기 위함
display(happiness_df.describe().T)
```

```python
# 기획: 대륙별 인생 선택의 자유도에 따른 행복지수 변화
# DataFrame을 간단하게 확인하기 위해 이외의 feature를 제거함
happiness_df = happiness_df.drop(labels=['country', 'generosity', 'social_support', 'gdp_per_capita', 'healthy_life_expectancy', 'perceptions_of_corruption'], axis=1)
display(happiness_df)
```

```python
import numpy as np
from scipy.stats import iqr

happiness_Q1 = happiness_df.describe().T.loc['happiness_score', '25%']
happiness_Q3 = happiness_df.describe().T.loc['happiness_score', '75%']
print(happiness_Q1, happiness_Q3)

iqr_value = iqr(happiness_df.happiness_score)
print(iqr_value)

lower_bound = happiness_Q1 - 1.5 * iqr_value
upper_bound = happiness_Q3 + 1.5 * iqr_value
print(f'정상치 범위: {lower_bound} ~ {upper_bound}')
```

```python
# 각 대륙 별 행복지수 평균 내림차순 정렬
hp_mean_df = happiness_df.groupby('region').mean('happiness_score',).reset_index()
hp_mean_df = hp_mean_df.sort_values(by='happiness_score', ascending=False)
hp_mean_df.reset_index(drop=True, inplace=True)
print('📈각 대륙 별 행복지수 평균 내림차순 정렬')
display(hp_mean_df)

print("=" * 78)

# 각 대륙 별 행복지수 최소값 정렬
hp_min_df = happiness_df.groupby('region').min('happiness_score').reset_index()
hp_min_df = hp_min_df.sort_values(by='happiness_score', ascending=True)
hp_min_df.reset_index(drop=True, inplace=True)
print('📈각 대륙 별 행복지수 최소값 정렬')
display(hp_min_df)

print("=" * 78)

# 각 대륙별 행복지수 최대값 정렬
hp_max_df = happiness_df.groupby('region').max('happiness_score').reset_index()
hp_max_df = hp_min_df.sort_values(by='happiness_score', ascending=False)
hp_max_df.reset_index(drop=True, inplace=True)
print('📈각 대륙 별 행복지수 최대값 정렬')
display(hp_max_df)

print("=" * 78)  
print("=" * 78)  
print("=" * 78)  

# 각 대륙별 삶의 선택 자유도 평균 내림차순 정렬
hp_mean_df = happiness_df.groupby('region').mean('freedom_to_make_life_choices',).reset_index()
hp_mean_df = hp_mean_df.sort_values(by='freedom_to_make_life_choices', ascending=False)
hp_mean_df.reset_index(drop=True, inplace=True)
print('📈각 대륙 별 삶의 선택 자유도 평균 내림차순 정렬')
display(hp_mean_df)

print("=" * 78)

# 각 대륙별 삶의 선택 자유도 최소값 정렬
hp_min_df = happiness_df.groupby('region').min('freedom_to_make_life_choices').reset_index()
hp_min_df = hp_min_df.sort_values(by='freedom_to_make_life_choices', ascending=True)
hp_min_df.reset_index(drop=True, inplace=True)
print('📈각 대륙 별 삶의 선택 자유도 최소값 정렬')
display(hp_min_df)

print("=" * 78)

# 각 대륙별 삶의 선택 자유도 최대값 정렬
hp_max_df = happiness_df.groupby('region').max('freedom_to_make_life_choices').reset_index()
hp_max_df = hp_min_df.sort_values(by='freedom_to_make_life_choices', ascending=False)
hp_max_df.reset_index(drop=True, inplace=True)
print('📈각 대륙 별 삶의 선택 자유도 최대값 정렬')
display(hp_max_df)

print("=" * 78)  
print("=" * 78)  
print("=" * 78)

# 각 대륙별 행복지수와 삶의 선택 자유도 표준편차
hp_std_df = happiness_df.groupby('region')[['happiness_score', 'freedom_to_make_life_choices']].std().reset_index()
print('📈각 대륙 별 행복지수와 삶의 선택 자유도 표준편차')
display(hp_std_df)

print("=" * 78)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 4, figsize=(20, 5))

hp_mean_df.plot(title='happiness_score_mean', x='region', y='happiness_score', fontsize='4', kind='bar', grid=True, ax=axes[0], ylabel="happiness_score")
hp_min_df.plot(title='happiness_score_min', x='region', y='happiness_score', fontsize='4', kind='bar', grid=True, ax=axes[1], ylabel="happiness_score")
hp_max_df.plot(title='happiness_score_max', x='region', y='happiness_score', fontsize='4', kind='bar', grid=True, ax=axes[2], ylabel="happiness_score")
hp_std_df.plot(title='happiness_score_std', x='region', y='happiness_score', fontsize='4', kind='bar', grid=True, ax=axes[3], ylabel="happiness_score")

fig, axes = plt.subplots(1, 4, figsize=(20, 5))

hp_mean_df.plot(title='freedom_to_make_life_choices_mean', x='region', y='freedom_to_make_life_choices', fontsize='4', kind='bar', grid=True, ax=axes[0], ylabel="freedom_to_make_life_choices")
hp_min_df.plot(title='freedom_to_make_life_choices_min', x='region', y='freedom_to_make_life_choices', fontsize='4', kind='bar', grid=True, ax=axes[1], ylabel="freedom_to_make_life_choices")
hp_max_df.plot(title='freedom_to_make_life_choices_max', x='region', y='freedom_to_make_life_choices', fontsize='4', kind='bar', grid=True, ax=axes[2], ylabel="freedom_to_make_life_choices")
hp_std_df.plot(title='freedom_to_make_life_choices_std', x='region', y='freedom_to_make_life_choices', fontsize='4', kind='bar', grid=True, ax=axes[3], ylabel="freedom_to_make_life_choices")
```
