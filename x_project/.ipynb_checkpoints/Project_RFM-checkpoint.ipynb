### 2. RFMÏóê Ï†êÏàò Í≥ÑÏÇ∞
- Recency: Ìï¥Îãπ Îç∞Ïù¥ÌÑ∞Ïóê ÏûàÎäî 'frequency of purchases' Î•º Í∞ÅÍ∞Å Ïà´ÏûêÎ°ú Î≥ÄÌòïÌï¥ÏÑú Ï†êÏàòÎ•º Î∂ÄÏó¨ ÌõÑ Í≥ÑÏÇ∞.
- Frequency: Ìï¥Îãπ Îç∞Ïù¥ÌÑ∞Ïóê ÏûàÎäî 'previous purchases' Ïóê 1Ïî© ÎçîÌï¥ÏÑú Í≥ÑÏÇ∞.
- Monetary: Ìï¥Îãπ Îç∞Ïù¥ÌÑ∞Ïóê ÏûàÎäî 'purchase amount' Î•º 'Monetary' Î°ú ÏÇ¨Ïö©.

```python
# Recency Í≥ÑÏÇ∞ Î∞è ÏÉùÏÑ±
# frequency of purchases Ïùò Í∞íÎì§ÏùÑ Ïà´ÏûêÎ°ú Î≥ÄÍ≤Ω ÌõÑ Recency featureÏóê ÏÉàÎ°≠Í≤å Îã¥ÎäîÎã§.

st_df['Recency'] = st_df['frequency of purchases'].replace({'Annually':1, 'Quarterly':2, 'Every 3 Months':3, \
                                                             'Monthly':4, 'Bi-Weekly':5, 'Fortnightly':6, \
                                                             'Weekly':7})
st_df
```

```python
# Monetary ÏÉùÏÑ±

st_df['Monetary'] = st_df['purchase amount']
st_df
```

```python
# Frequency feature ÏÉùÏÑ±

st_df['Frequency'] = st_df['previous purchases'] + 1
st_df
```

```python
m_st_df = st_df[['id','Monetary']]
m_st_df
```

```python
r_st_df = st_df[['id','Recency']]
r_st_df
```

```python
f_st_df = st_df[['id', 'Frequency']]
f_st_df
```

```python
# r_st_df, f_st_df, m_st_dfÏùÑ rfm_st_dfÎ°ú Î≥ëÌï©

rfm_st_df = pd.merge(r_st_df, f_st_df, on='id')
rfm_st_df = pd.merge(rfm_st_df, m_st_df, on='id')

rfm_st_df
```

```python
from sklearn.preprocessing import MinMaxScaler

rfm_scale = MinMaxScaler()
rfm_scale = rfm_scale.fit_transform(rfm_st_df[['Recency', 'Frequency', 'Monetary']])
```

```python
rfm_st_df[['Recency', 'Frequency', 'Monetary']] = pd.DataFrame(rfm_scale, columns=['Recency', 'Frequency', 'Monetary'])
rfm_st_df
```

```python
rfm_st_df.index += 1
```

```python
rfm_st_df.loc[:, 'TotalScore'] = rfm_st_df['Recency'] + rfm_st_df['Frequency'] + rfm_st_df['Monetary']
rfm_st_df
```

```python
st_rfm_df = st_df.merge(rfm_st_df[['id', 'TotalScore']], on='id')
st_rfm_df
```

### 3. 'Total Score' Î°ú ÌöåÏõêÎì§ÏóêÍ≤å Îì±Í∏â Î∂ÄÏó¨

```python
import numpy as np

l1, l2, l3, l4 = np.percentile(st_rfm_df['TotalScore'], [30, 55, 75, 90])
print(l1, l2, l3, l4)
```

```python
def get_level(x):
    if x <= l1:
        return 5
    if x <= l2:
        return 4
    if x <= l3:
        return 3
    if x <= l4:
        return 2
    return 1
```

```python
st_rfm_df['level'] = st_rfm_df['TotalScore'].apply(get_level)
st_rfm_df
```

```python
st_rfm_df['level'] = st_rfm_df['level'].replace([5, 4, 3, 2, 1], 
                                                ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP'])
```

```python
# ÎÇòÏù¥Î•º Í∑∏Î£πÌôîÌïòÎäî Ìï®Ïàò
def categorize_age(age):
    if age < 20:
        return '10s'
    elif age < 30:
        return '20s'
    elif age < 40:
        return '30s'
    elif age < 50:
        return '40s'
    elif age < 60:
        return '50s'       
    else:
        return '60s'
```

```python
st_rfm_df['age_group'] = st_rfm_df['age'].apply(categorize_age)
st_rfm_df
```
