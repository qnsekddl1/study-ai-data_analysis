### 2.ÏÑ±Î≥Ñüöπüö∫
ÏÑ±Î≥ÑÏóê Îî∞Î•∏ Í∞úÏù∏Ïùò Í≤ΩÌóòÍ≥º ÏÉùÍ∞ÅÏùò Ï∞®Ïù¥Îäî ÎÑàÎ¨¥ÎÇòÎèÑ ÎãπÏó∞Ìïú Í≤∞Í≥º ÏûÖÎãàÎã§.  
Í∑∏Î†áÎã§Î©¥, ÏÑ±Î≥ÑÎ°ú Ïù∏Ìïú ÏÉùÍ∞ÅÏùò Ï∞®Ïù¥Îäî Í≥ß, Ïä§Ìä∏Î†àÏä§ Ï¶ùÍ∞ÄÏóê ÎåÄÌïú Ï∞®Ïù¥ÎèÑ ÏûàÏùÑ Ïàò ÏûàÏùÑÍ±∞Îùº ÏÉùÍ∞ÅÏù¥ Îì≠ÎãàÎã§.  
Í∑∏Î†áÍ∏∞Ïóê Ïù¥ Î∂ÄÎ∂ÑÏóêÏÑúÎäî ÏÑ±Î≥ÑÏóê ÎåÄÌïú Ïä§Ìä∏Î†àÏä§ Ï¶ùÍ∞Ä Ï∞®Ïù¥Î•º Ï°∞ÏÇ¨Ìï¥Î≥¥ÎèÑÎ°ù ÌïòÍ≤†ÏäµÎãàÎã§.

#### üñäÔ∏èÏÑ§Î¨∏ Ï°∞ÏÇ¨Ïóê Ï∞∏Ïó¨Ìïú Ïù∏ÏõêÎì§Ïùò ÏÑ±ÎπÑÎ•º Ï°∞ÏÇ¨Ìï¥Î≥¥Ïûê.

```python
# ÏÑ±Î≥ÑÏóê Îî∞Î•∏ Ïù∏Ïõê Ïàò Í≥ÑÏÇ∞
gender_counts = mh['Gender'].value_counts()

plt.rcParams['font.family'] = 'Malgun Gothic'

colors = ['#00aaff', '#ff6347']

wedgeprops = {'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

explode = (0, 0.08)

# ÏõêÌòï Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=colors, wedgeprops=wedgeprops, explode=explode)
plt.title('ÏÑ§Î¨∏Ï°∞ÏÇ¨Ïóê Ï∞∏Ïó¨Ìïú Ïù∏ÏõêÎì§Ïùò ÏÑ±ÎπÑ')
plt.axis('equal')  # ÏõêÌòï Í∑∏ÎûòÌîÑÎ•º ÏõêÏúºÎ°ú ÏÑ§Ï†ï
plt.show()
```

üìùÏÑ§Î¨∏ Ï°∞ÏÇ¨Ïóê Ï∞∏Ïó¨Ìïú Ïù∏ÏõêÎì§Ïùò ÏÑ±ÎπÑÍ∞Ä Ïó¨ÏÑ± 18.0%, ÎÇ®ÏÑ±Ïùò ÎπÑÏú®Ïù¥ 82.0% Î°ú, ÏïΩ 4.5Î∞∞Ï†ïÎèÑ Îçî ÎßéÏäµÎãàÎã§.

#### üñäÔ∏èÏä§Ìä∏Î†àÏä§ Ï¶ùÍ∞ÄÏôÄ ÏÑ±Î≥ÑÏùò Í¥ÄÍ≥ÑÏóê ÎåÄÌï¥ Ï°∞ÏÇ¨Ìï¥Î≥¥Ïûê.

```python
plt.rcParams['font.family'] = 'Malgun Gothic'

plt.figure(figsize=(10, 7))
sns.countplot(x='Growing_Stress', data=mh, hue='Gender',  palette = 'rocket')
plt.title('Ïä§Ìä∏Î†àÏä§ÏôÄ ÏÑ±Î≥ÑÏùò Í¥ÄÍ≥Ñ')
plt.xlabel('Ïä§Ìä∏Î†àÏä§ Ï¶ùÍ∞Ä')
plt.ylabel('')
plt.show()
```
