```python
import pandas as pd
import numpy as np

np = pd.read_csv('./datasets/Netflix subscription fee Dec-2021.csv')
np
```

```python
np.info()
```

```python
np.isna().sum()
```

```python
# 컬럼명 변경
np.columns = ['C_code', 'Country', 'Total Size', 'TV Shows', 'Movies', 'Basic', 'Standard', 'Premium']
np
```

```python
# 각 요금제 가입 시 콘텐츠 당 가격
np['Basic Price'] = np['Basic'] / np['Total Size']
np['Standard Price'] = np['Standard'] / np['Total Size']
np['Premium Price'] = np['Premium'] / np['Total Size']
np
```

```python
# Basic Price가 가장 낮은 가격을 가지는 상위 5개 국가 선택

sorted_bp_a = np.sort_values(by="Basic Price")
ch_bp = sorted_bp_a.head(5)

ch_bp
```

```python
# Standard Price 가장 낮은 가격을 가지는 상위 5개 국가 선택

sorted_sp_a = np.sort_values(by="Standard Price")
ch_sp = sorted_sp_a.head(5)

ch_sp
```

```python
# Premium Price가 가장 낮은 가격을 가지는 상위 5개 국가 선택

sorted_pp_a = np.sort_values(by="Premium Price")
ch_pp = sorted_pp_a.head(5)

ch_pp
```

```python
# Basic Price가 가장 높은 가격을 가지는 상위 5개 국가 선택

sorted_bp_b = np.sort_values(by="Basic Price", ascending=False)
ex_bp = sorted_bp_b.head(5)

ex_bp
```

```python
# Standard Price가 가장 높은 가격을 가지는 상위 5개 국가 선택

sorted_sp_b = np.sort_values(by="Standard Price", ascending=False)
ex_sp = sorted_sp_b.head(5)

ex_sp
```
