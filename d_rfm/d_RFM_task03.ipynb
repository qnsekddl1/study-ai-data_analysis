### RFM 분석 실습
- Recency: 얼마나 최근에 구매했는가
- Frequency: 얼마나 자주 구매했는가
- Monetary: 얼마나 많은 금액을 지출했는가

### 고객 분석

> voice_no: 송장 번호. 명사 같은. 각 작업에 고유하게 할당된 문자 'I'와 6자리 정수의 조합  
customer_id: 고객번호. 명사 같은. 각 작업에 고유하게 할당된 문자 'C'와 6자리 정수의 조합  
gender: 고객 성별의 문자a열 변수  
age: 고객 연령의 양의 정수 변수  
category: 구매한 상품 카테고리의 문자열 변수 
quantity: 거래당 각 제품(항목)의 수량  
price: 단가. 숫자. 단위당 제품 가격은 터키 리라(TL)  
payment_method: 거래에 사용된 결제 방법(현금, 신용 카드 또는 체크 카드)의 문자열 변수  
invoice_date: 송장 날짜. 거래가 발생한 날  
shopping_mall: 거래가 이루어진 쇼핑몰 이름의 문자열 변수

```python
cs_df.info()
```

```python
cs_df = cs_df.sort_values('date', ascending=True)
```

```python
category_price_sum = cs_df.groupby('category')['price'].sum().reset_index()
category_price_sum
```

```python
cs_df
```

```python
# 가장 최근에 구매한 날짜('2023-03-08') 로
cs_df['Recency'] = pd.to_datetime('2023-03-08') - pd.to_datetime(cs_df['date'])
cs_df['Recency'] = cs_df['Recency'].dt.days
cs_df
```

```python
cs_df = cs_df.drop(['date'], axis=1)
```

```python
# 'quantity', 'price' 가 음수인 것

print((cs_df['quantity'] < 0).sum())
print((cs_df['price'] < 0).sum())
```

```python
cs_df[cs_df['quantity'] >= 0]
```

```python
rf_df = cs_df.groupby('id').agg(Recency=('Recency', 'min'), Frequency=('id', 'count'))
rf_df
```

```python
m_df = cs_df.copy()
m_df.loc[:, 'Monetary'] = m_df['quantity'] * m_df['price']
m_df
```

```python
m_df = m_df.groupby('id')[['Monetary']].sum()
m_df
```

```python
rf_df = rf_df.reset_index()
rf_df
```

```python
m_df = m_df.reset_index()
m_df
```

```python
rfm_df = rf_df.merge(m_df, on='id')
rfm_df
```

```python
rfm_df.iloc[rfm_df[rfm_df['Monetary'] < 0].index] = 0
rfm_df[rfm_df['Monetary'] < 0]
```

```python
rfm_df
```

```python
from sklearn.preprocessing import MinMaxScaler

rfm_scale = MinMaxScaler()
rfm_scale = rfm_scale.fit_transform(rfm_df[['Recency', 'Frequency', 'Monetary']])
```

```python
rfm_df[['Recency', 'Frequency', 'Monetary']] = pd.DataFrame(rfm_scale, columns=['Recency', 'Frequency', 'Monetary'])
rfm_df
```

```python
rfm_df['Recency'] = 1 - rfm_df['Recency']
```

```python
rfm_df
```

```python
rfm_df.loc[:, 'TotalScore'] = rfm_df['Recency'] + rfm_df['Frequency'] + rfm_df['Monetary']
rfm_df
```

```python
cs_rfm_df = cs_df.merge(rfm_df[['id', 'TotalScore']], on='id')
cs_rfm_df
```

```python
cs_rfm_df.drop(columns=['Recency'], inplace=True)
```

```python
cs_rfm_df
```

```python
import numpy as np

l1, l2, l3, l4 = np.percentile(cs_rfm_df['TotalScore'], [20, 45, 70, 92])
print(l1, l2, l3, l4)
```

```python
def get_level(x):
    if x <= l1:
        return 5
    if x <= l2:
        return 4
    if x <= l3:
        return 3
    if x <= l4:
        return 2
    return 1
```

```python
cs_rfm_df['Level'] = cs_rfm_df['TotalScore'].apply(get_level)
cs_rfm_df
```

```python
cs_rfm_df['Level'] = cs_rfm_df['Level'].replace([5, 4, 3, 2, 1], 
                                                ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP'])
```

```python
cs_rfm_df
```

```python
# 나이를 그룹화하는 함수
def categorize_age(age):
    if age < 20:
        return '10s'
    elif age < 30:
        return '20s'
    elif age < 40:
        return '30s'
    elif age < 50:
        return '40s'
    elif age < 60:
        return '50s'
    else:
        return '60s'
```

```python
cs_rfm_df['age_group'] = cs_rfm_df['age'].apply(categorize_age)
cs_rfm_df
```

### 📝마케팅 전략 보고서

1. 제일 먼저 전 쇼핑몰의 연령대별 회원 수를 파악했을 때, 20대부터 ~ 60대까지 고르게 분포하고 있고,  
언뜻 봐선 10대가 적어 보이겠지만 최연소 회원의 나이가 18살인 것을 감안한다면 10대 역시 고르게 분포해있다고 볼 수 있다.  
이에 따라 앞으로도 한 연령대에 치우치지 않게 운영 해나가야 할 것이다.

2. 전 쇼핑몰의 결제 방식의 비율을 확인 해보았다.  현금, 신용 카드, 체크 카드 순으로 많은 것을 알 수 있다.  
그리고 연령대별 결제 방식을 확인 해보았을 때, 현금 결제와 체크 카드를 가장 많이 한 나이대는 20대, 신용 카드 결제를 가장 많이 한 나이대는 30대이다.  
하지만 결제하는 데 신용 카드, 체크 카드와 같이 카드 결제를 할 경우 카드 사에 결제수수료를 지급해야야하는 마이너스 요인이 있으니 현금 사용을 유도하는 방향의 캠페인이 있다면 좋을 듯하다.

3. 연령대별 구매 금액의 총합을 확인 해보았다.  
우선 40대 13,376,514 리라로 가장 높았고, 이어서 20대가 13,324,658 리라였고 30대가 13,245,827 리라였다.  
이에 따라 주력 고객층인 2049 고객들의 취향과 트랜드에 맞는 제품들을 판매한다면 기존의 2049 고객 유지와 나아가 신규 고객 유치에도 도움이 될 것으로 파악된다.  
하지만 너무 2049 고객들을 위한 편향적인 제품들로만 판매들이 된다면 주력 고객 이외 고객들의 불만이 커질 수도 있으니
이외 고객들의 흥미를 유발할 수 있는 제품들도 적절히 판매하는 것이 좋을 것이다.  

4. 카테고리별 팔린 수량, 카테고리별 판매한 총금액을 확인했을 때, 의류가 다른 카테고리들에 비해 압도적으로 많은 수량과 금액이 놓은 것으로 나타났다.  
위의 그래프 중 '나이대별 구매한 카테고리' 를 확인해 보면, 전 연령층에서 의류 부문이 가장 많이 팔린 것으로 확인된다.
하지만 책, 식료품, 기념품 부문은 너무나도 낮은 판매율을 보인다.  
이에 따라 의류 부문에 대한 개선 사항은 없으나, 가장 판매율이 부진한 책, 식료품, 기념품 부문에 대해선 캠페인을 진행해 고객들의 구매를 유발하여야 할 듯 하다.

5. 전 쇼핑몰의 대상으로 회원의 성비를 조사 했을 때, 여성 59.8%, 남성 40.2% 로 여성이 높은 비율을 보였다.  
그리고 당연히 여성의 성별이 높기 때문에 카테고리별 성별의 구매한 총금액을 조사 했을 때도 모든 카테고리에서도 여성의 총구매 금액이 높았다.

### 📌결론  
연령대별 회원의 수는 적절히 분포해 있고 결제 방식에 대해선 개선 사항이 없지만, 추후 동향을 살펴야 한다.
