### RFM 분석 실습
- Recency: 얼마나 최근에 구매했는가
- Frequency: 얼마나 자주 구매했는가
- Monetary: 얼마나 많은 금액을 지출했는가

### 고객 분석
- InvoiceNo: 거래 번호
- StockCode: 상품 코드
- Description: 상품명 
- InvoiceDate: 거래 날짜
- UnitPrice: 상품 단가
- CustomerID: 고객 번호
- Country: 국가명

Frequency가 없으면 Quantity로 대체

```python
import pandas as pd
import numpy as np

ci_df = pd.read_csv('./datasets/customer_invoice_data.csv')
ci_df
```

```python
ci_df.info()
```

```python
# 컬럼명 변경
ci_df.columns = ['OrderID','StockCode', 'Information', 'Quantity', 'Date', 'Price', 'ID', 'Country']
ci_df
```

```python
# StockCode 열 삭제
ci_df.drop(columns=['StockCode', 'Information'], inplace=True)
ci_df
```

```python
# DataFrame의 중복 행 검사
ci_df.duplicated().sum()
```

```python
# DataFrame의 결측치 검사
ci_df.isna().sum()
```

```python
ci_df['ID'].value_counts()
```

```python
# Monetary 구함

ci_df_x = ci_df.groupby('ID').sum('Price').reset_index()
ci_df_x.drop(columns=['Quantity'], inplace=True)
ci_df_x
```

```python
ci_df
```

```python
ci_df = ci_df[ci_df['Quantity'] > 0]
ci_df
```

```python
# Date 형식 변경
ci_df['Date'] = pd.to_datetime(ci_df['Date'])
ci_df.sort_values(['Date'], ascending=False)

# Date 형식 변경 = 시분초 제거
ci_df['Date'] = ci_df['Date'].dt.strftime('20'+'%y-%m-%d')
ci_df
```

```python
ci_df['Date'] = pd.to_datetime(ci_df['Date'])
```

```python
# Recency
ci_df['Recency'] = (pd.to_datetime('2011-12-09') - ci_df['Date']).dt.days
```

```python
ci_df.duplicated().sum()
```

```python
ci_df.drop_duplicates(inplace=True)
```

```python
ci_df
```

```python
ci_df.reset_index()
```

```python
ci_df.isnull().sum()
```

```python
ci_df = ci_df.dropna(subset=['ID'])
```

```python
ci_df
```

```python
ci_df = ci_df.reset_index()
ci_df
```

```python
# Recency
rfm_r = ci_df.groupby('ID')['Recency'].min().reset_index()
rfm_r.columns = ['ID', 'Recency']

# Frequency
rfm_f = ci_df['ID'].value_counts().reset_index()
rfm_f.columns = ['ID', 'Frequency']

# Monetary
rfm_m = ci_df.groupby('ID')['Price'].sum().reset_index()
rfm_m.columns = ['ID', 'Monetary']

rfm = pd.merge(rfm_r, rfm_f, on='ID')
rfm = pd.merge(rfm, rfm_m, on='ID')

rfm
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
rfm[['Recency', 'Frequency', 'Monetary']] = \
                                    scaler.fit_transform(rfm[['Recency', 'Frequency', 'Monetary']])
```

```python
# 최근일 수록 숫자가 더 작게 나오기 때문에, 1을 빼준 값으로 점수를 계산한다.
rfm['Recency'] = 1 - rfm['Recency']
rfm
```

```python
# RFM 점수 계산
rfm['TotalScore'] = rfm[['Recency', 'Frequency', 'Monetary']].sum(axis=1)
rfm
```

```python
ci_rfm_df = pd.merge(ci_df, rfm, on='ID')
ci_rfm_df = ci_rfm_df.rename(columns={'Recency_y': 'Recency'})
ci_rfm_df.drop(columns=['Recency_x'], inplace=True)
ci_rfm_df
```

```python
import numpy as np

l1, l2, l3 = np.percentile(ci_rfm_df['TotalScore'], [25, 55, 90])
print(l1, l2, l3)
```

```python
def get_level(x):
    if x <= l1:
        return 4
    if x > l1 and x <= l2:
        return 3
    if x > l2 and x <= l3:
        return 2
    return 1
```

```python
ci_rfm_df['Level'] = ci_rfm_df['TotalScore'].apply(get_level)
ci_rfm_df
```

```python
ci_rfm_df['Level'] = ci_rfm_df['Level'].replace([4, 3, 2, 1],['Bronze', 'Silver', 'Gold', 'VIP'])
```

```python
ci_rfm_df
```

```python
ci_rfm_df = ci_rfm_df.groupby('ID')[['OrderID', 'Quantity', 'Date', 'Price', 'Country', 'TotalScore', 'Level']].first().reset_index()
ci_rfm_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

order_columns = ['Bronze', 'Silver', 'Gold', 'VIP']

sns.countplot(x='Level', data=ci_rfm_df, palette='muted', order=order_columns)
```
